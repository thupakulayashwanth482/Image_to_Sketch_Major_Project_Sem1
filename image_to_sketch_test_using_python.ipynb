```python
import cv2
from matplotlib import pyplot as py
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("wallpaperflare.com_wallpaper.jpg")

#give your own image as input for testing after uploading it on to colab
```

```python
#original image section

img_shrinked = cv2.resize(image,(350, 300), interpolation = cv2.INTER_AREA)
print("ORGINAL IMAGE")
cv2_imshow(img_shrinked)
```

```python
#grey image converstion section

grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img_shrinked = cv2.resize(grey_img,(350, 300), interpolation = cv2.INTER_AREA)
```

```python
print("grey image convertion")
cv2_imshow(img_shrinked)
```

```python
#inverting the converted grey image section.......

invert = cv2.bitwise_not(grey_img)
```

```python
img_shrinked = cv2.resize(invert,(350, 300), interpolation = cv2.INTER_AREA)
print("INVERTED IMAGE")
cv2_imshow(img_shrinked)
```

```python
#blurring the inverted image section ..............

blur = cv2.GaussianBlur(invert,(21,21),0)
```

```python
img_shrinked = cv2.resize(blur,(350, 300), interpolation = cv2.INTER_AREA)
print("BLUR IMAGE")
cv2_imshow(img_shrinked)
```

```python
#inverting the blurred image section ...

inverted_blur = cv2.bitwise_not(blur)
```

```python
img_shrinked = cv2.resize(inverted_blur,(350, 300), interpolation = cv2.INTER_AREA)
print("INVERTED BLUR IMAGE")
cv2_imshow(img_shrinked)
```

```python
#sketch section .........

sketch = cv2.divide(grey_img,inverted_blur,scale=256.0)
```

```python
img_shrinked = cv2.resize(sketch,(350, 300), interpolation = cv2.INTER_AREA)
print("SKETCH IMAGE")
cv2_imshow(img_shrinked)
```

```python
cv2.imwrite("sketch.png",sketch)
```
